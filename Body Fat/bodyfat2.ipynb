```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns #import libraries
```

```python
data = pd.read_csv('bodyfat.csv')  #read csv
```

```python
new_data_process = data.set_index('Age')  #assign new variable and set index to 'Age'
```

```python
new_data_dummy = pd.get_dummies(new_data_process) #get dummies from new_data_process and assign it to a vairable
new_data_dummy = pd.get_dummies(new_data_process, drop_first=True)  #same as above but drops the first value
new_data_dummy.head() #shows
```

```python
input_new = new_data_dummy[['Weight','Height','BodyFat']] #assign new columns to the data
input_new.head()  #print
```

```python
from sklearn.preprocessing import StandardScaler  #import libraries
scaler = StandardScaler() #assign scaler to the function StandardScaler()
scaler.fit(input_new) #standardizes and fits the value from input_new
input_new = scaler.transform(input_new) #assign input_new and transforms it
```

```python
input_new #prints
```

```python
import pickle #import library
filename = 'reg_model_save.sav' #
loaded_model = pickle.load(open(filename,'rb')) #load model and open the filename 'rb"
```

```python
import statsmodels.api as sm  #import liibrary for statistical analysis and modeling
input_new = sm.add_constant(input_new)  #adds a constant term to "input_new" so we can estimate 
predict_output = loaded_model.predict(input_new)  #predict target variable for the new input data
predict_output = pd.DataFrame({'Predicted Body Fat':predict_output})  #creates new df called predict_output to store new values
```

```python
predict_output.head() #
```

```python
new_predict = pd.concat([data, predict_output],axis=1)  #concatenantes data and predict_output
```
